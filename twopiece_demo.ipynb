# Twopiece Library Demo

This notebooks illustrates the main functionality of the **[twopiece](https://github.com/quantgirluk/twopiece)** Python library.

- **Download:** https://pypi.org/project/twopiece/
- **Source Code:** https://github.com/quantgirluk/twopiece

The twopiece library provides a Python implementation of the family of Two Piece distributions.

**Definition.** Let $f: \mathbb{R} \mapsto \mathbb{R}_{+}$ be a unimodal symmetric (about 0) probability density function (pdf) from the [location-scale family](https://en.wikipedia.org/wiki/Location%E2%80%93scale_family), possibly including a shape parameter $\delta$. Then, the pdf of a member of the two-piece family of distributions is given by

$$
s(x) := s\left(x; \mu,\sigma_1,\sigma_2, \delta\right) =
  \begin{cases}
\dfrac{2}{\sigma_1+\sigma_2}f\left(\dfrac{x-\mu}{\sigma_1};\delta\right), \mbox{if } x < \mu, \\
\dfrac{2}{\sigma_1+\sigma_2}f\left(\dfrac{x-\mu}{\sigma_2};\delta\right), \mbox{if } x \geq \mu. \\
\end{cases}
$$

## 3. Creating an Instance

The **twopiece** library supports 4 distributions with three parameters (location, sigma1, sigma2) and 2 distributions with four parameters (location, sigma1, sigma2, shape).

#### Three Parameters

- two-piece normal is **tpnorm**
- two-piece Laplace is **tplaplace**
- two-piece Cauchy is **tpcauchy**
- two-piece logistic **tplogistic**

#### Four Parameters
- two-piece t Student is **tpstudent**
- two-piece exponential power also called two-piece generalised normal is **tpgennorm**

We can create an instance by simply definign the parameters.

### Example Two-piece normal: **tpnorm**

````python
dist = tpnorm(loc=0.0, sigma1=1.0, sigma2=1.5)
````

### Example Two-piece Student t: **tpstudent**

````python
dist = tpstudent(loc=1.0, sigma1=0.5, sigma2=0.8, shape=3.0)
````

## 4. Methods

We provide the following methods for each instance

- probability density function **pdf**
- cumulative distribution function **cdf**
- quantile function **ppf**
- random generation **random_sample**

We can evaluate **pdf, cdf, ppf** on a single value or an array type object.

### Example Two-piece normal: **tpnorm**

````python
dist = tpnorm(loc=0.0, sigma1=2.0, sigma2=1.5)
dist.pdf(0)
````

````python
dist.pdf([0.0,0.25,0.5])
````

````python
x = arange(-5, 5, 0.1)
y = dist.pdf(x)
plt.plot(x, y)
plt.show()
````

## 5. Display Function

The **twopiece** library function **display_tpdist** which displays plots for any of the supported distributions.
This function requieres the following parameters:

- ***loc***: location parameters (optional, default=0.0)
- ***sigma1***: left scale parameters (optional, default=1.0)
- ***sigma2***: location parameters (optional, default=0.0)
- ***shape***: shape parameters (optional, default=3.0) This one is required when we want to illustrate the 3 parameters instances.
- ***tpdist***: string 'tpnorm', 'tplaplace', 'tpcauchy', 'tpstudent', 'tpgennorm', 'All' (to display all).
- ***show***: string 'pdf', 'cdf', 'ppf', 'random_sample', 'All' (to show all).



````python
display_tpdist(loc=0.5, sigma1=1.5, sigma2=2.5, shape=3.0, tpdist='tpgennorm', show='random_sample');
````

````python
display_tpdist(sigma1=1.0, sigma2=2.5, shape=3.0, tpdist='All', show='pdf');
````

````python
display_tpdist(sigma1=1.0, sigma2=2.5, shape=3.0, tpdist='All', show='All');
````
